```python
from datetime import datetime

def no_of_records(num): # initial value is string
    valid_flag = True
    if not num.isdigit():
        print("Number of records must be an integer.")
        valid_flag = False
    num = int(num) # typecast to integer
    if num <= 0 or num > 30:
        print("Number of records must be between 1-30.")
        valid_flag = False
    return valid_flag


def subject_code(code):
    valid_flag = True
    if not code.isdigit() or len(code) != 4:
        print("Subject code must be a 4 digit number")
        valid_flag = False
    return valid_flag

def name(name):
    valid_flag = True
    '''
    if not name.isalnum():
        print("Name must contain only alphanumeric characters.")
        valid_flag = False
        '''
    if len(name) < 3 or len(name) > 30:
        print("Name must be between 3-30 characters long.")
        valid_flag = False
    return valid_flag

def sub_type(sub):
    valid_flag = True
    try:
        sub = str(sub)
        if sub.upper() not in ('A', 'P'):
            print("Subject type must be either 'A' or 'P'.")
            valid_flag = False
    except ValueError:
        print("Subject type must be either 'A' or 'P'.")
        valid_flag = False
    return valid_flag

def p1_len(p1):
    valid_flag = True
    try:
        p1 = float(p1)
        if p1 not in (2.0, 2.5, 3.0):
            print("Duration of paper 1 must be either 2.0, 2.5 or 3.0.")
            valid_flag = False
    except ValueError:
        print("Paper 1 must be a floating point number (e.g., 2.0)")
        valid_flag = False
    return valid_flag

def p2_len(p2):
    valid_flag = True
    try:
        p2 = float(p2)
        if p2 not in (2.0, 2.5, 3.0):
            print("Duration of paper 2 must be either 2.0, 2.5 or 3.0.")
            valid_flag = False
    except ValueError:
        print("Paper 2 must be a floating point number (e.g., 2.0)")
        valid_flag = False
    return valid_flag

def c_date(date): # format DDMMYY
    valid_flag = True
    # type check
    if not date.isdigit() or len(date) != 6:
        print("Date must contain 6 digits in the format DDMMYY.")
        valid_flag = False
    else:
        # check for day
        day = int(date[0:2])
        if day < 1 or day > 31:
            print("Day out of range.")
            valid_flag = False
        # check for month
        month = int(date[2:4])
        if month < 1 or month > 12:
            print("Month out of range.")
            valid_flag = False
        # check for year
        year = date[4:]
        local_year = str(datetime.today().year)[2:]
        if year != local_year:
            print("Year is incorrect.")
            valid_flag = False
    return valid_flag

def prac_len(prac):
    valid_flag = True
    try:
        prac = float(prac)
        if prac < 10.0 or prac > 40.0:
            print("Duration of practical must be between 10.0-40.0")
            valid_flag = False
    except ValueError:
        print("Duration of practical must be a floating point number (e.g., 20.0).")
        valid_flag = False
    return valid_flag
```
